```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')
from sklearn import linear_model
```

```python
ecommerce = pd.read_csv('/content/ecommerce-product-reviews.csv')
ecommerce
```

```python
ecommerce.head()
```

```python
LE_review = LabelEncoder()
ecommerce["review"] =LE_review.fit_transform(ecommerce['review'])
```

```python
LE_label = LabelEncoder()
ecommerce["label"] =LE_label.fit_transform(ecommerce['label'])
```

```python
input = ecommerce.drop(['label'],axis=1)
input
```

```python
target = ecommerce['label']
target
```

```python
ecommerce.head()
```

```python
model = linear_model.LinearRegression()
model
```

```python
model.fit(input,target)
```

```python
model.score(input,target)
```
