```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
import warnings
warnings.filterwarnings('ignore')
from sklearn import linear_model
```

```python
ecommerce = pd.read_csv('/content/ecommerce-product-reviews.csv')
ecommerce
```

```python
ecommerce.head()
```

```python
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
model
```

```python
# 2. Split data (X = reviews, y = labels)
X_train, X_test, y_train, y_test = train_test_split(ecommerce['review'], ecommerce['label'], test_size=0.2)
X_train, X_test, y_train, y_test
```

```python
model = make_pipeline(TfidfVectorizer(), linear_model.LogisticRegression())
model
```

```python
#  Train and Score
model.fit(X_train, y_train)
```

```python
print(model.score(X_test, y_test))
```
